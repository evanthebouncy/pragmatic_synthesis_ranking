```python
import numpy as np

global M, H, U

# function to read file and return lines
def read_file(filename):
    with open(filename) as f:
        return [i.strip() for i in f.readlines()]

class tdata:
    def __init__(self, h_index, utterances, h_order):
        self.h_index = h_index
        self.utterances = utterances
        self.h_order = h_order
    
    def __str__(self):
        return "h_index: " + str(self.h_index) + " utterances: " + str(self.utterances) + " h_order: " + str(self.h_order)

# function to read the training data
def read_training_data(filename):
    lines = read_file(filename)
    t_data = []
    for line in lines:
        line = line.strip()
        if line:
            h_index, utterances, h_order = line.split(';')
            h_index = int(h_index)
            # remove [ and ] from utterances
            utterances = utterances[1:-1]
            utterances = [int(j) for j in utterances.split(',')]
            # remove [ and ] from h_order
            h_order = h_order[1:-1]
            h_order = [int(j) for j in h_order.split(',')]
            t_data.append(tdata(h_index, utterances, h_order))
    return t_data

def loadData(mm_file, regex_file, sentences_file):
    
    mm = np.load(mm_file)
    regexes = read_file(regex_file)
    sentences = read_file(sentences_file)

    return mm, regexes, sentences

def is_consistent(h_index, utterances):
    for i in range(len(utterances)):
        if M[utterances[i]][h_index] == 1:
            return True
    return False
```

```python
M, H, U = loadData('../data/m350/mm.npy', '../data/m350/regexes.txt', '../data/m350/sentences.txt')

training_data = read_training_data('../data/m350/training_data.txt')
# print([str(i) for i in training_data[:3]])

# fixing the training data
for t in training_data[1:]:
    h_consistent = [h for h in t.h_order if is_consistent(h, t.utterances)]
    print([U[i] for i in t.utterances])
    print([H[i] for i in t.h_order])
    print([H[i] for i in h_consistent])
    break
    
```
